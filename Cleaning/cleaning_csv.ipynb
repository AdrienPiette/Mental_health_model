```python
import pandas as pd

def clean_data(df):
    # Replace missing values with the most common value of each column in: 'Mental_Health_Condition', 'Physical_Activity'
    df = df.fillna({'Mental_Health_Condition': df['Mental_Health_Condition'].mode()[0], 'Physical_Activity': df['Physical_Activity'].mode()[0]})
    return df

# Loaded variable 'df' from URI: c:\Users\pieta\OneDrive\Bureau\Mental Health Model\Mental_health_model\Data\Impact_of_Remote_Work_on_Mental_Health.csv
df = pd.read_csv(r'c:\Users\pieta\OneDrive\Bureau\Mental Health Model\Mental_health_model\Data\Impact_of_Remote_Work_on_Mental_Health.csv')

df_clean = clean_data(df.copy())
df_clean.head()

# save the cleaned data to a new csv file

df_clean.to_csv(r'c:\Users\pieta\OneDrive\Bureau\Mental Health Model\Mental_health_model\Data\Impact_of_Remote_Work_on_Mental_Health_cleaned.csv', index=False)
```

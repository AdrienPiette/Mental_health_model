```python
import pandas as pd

def clean_data(data):
    # Replace missing values with the most common value of each column in: 'Mental_Health_Condition', 'Physical_Activity'
    data = data.fillna({'Mental_Health_Condition': data['Mental_Health_Condition'].mode()[0], 'Physical_Activity': data['Physical_Activity'].mode()[0]})
    return data


df_clean = clean_data(data.copy())
df_clean.head()

```
